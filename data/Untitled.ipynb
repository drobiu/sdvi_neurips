```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

$$ y_i = \Sigma^{P4}_{d=1}β_d x_{i,d} + ϵ_i $$ where $ ϵ_i ∼ N(0, 1)$, $x_{i,d} ∼ N (0, 1)$ and $ β = [1.5, 1.5, 0.3, 0.1] $

```python
beta = [1.5, 1.5, 0.3, 0.1]
```

```python
def generate_data(n):
    xs = []
    ys = []
    
    for i in range(n):
        y = 0
        x = []
        e = np.random.normal()

        for b in beta:
            x_current = np.random.normal()
            x.append(x_current)
            y += b * x_current + e
        
        xs.append(x)
        ys.append(y)
        
    return np.array(xs), np.array(ys)
```

```python
generate_data(100)
```
